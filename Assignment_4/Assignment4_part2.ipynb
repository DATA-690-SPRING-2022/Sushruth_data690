## Cleanse the data - use apply() function




> ### Cleaning YEAR

```
```








````python
data['Census Year'].unique()
````

````python
def clean_year(a):
    return int(a.replace("*",""))
````

````python
data["Year"] = data["Census Year"].apply(clean_year)
data
````



> ### Cleaning population column



````python
data['Total Population'].unique()
````

````python
def clean_population(a):
    return int(a.replace(",",""))
````

````python
data["Population"] = data["Total Population"].apply(clean_population)
data
````



> ### Cleaning Cost column



````python
def clean_cost(data):
    data = data.replace(" ","")
    data = data.replace("\n","")
    data = data.replace("*","")
    data = data.replace(",","")
    data = data.replace("$","")

    if "Billion" in data:
        if "." in data:
            var = "00000000"
            data = data.replace("Billion", var)
            data = data.replace(".", "")
        else:
            var = "000000000"
            data = data.replace("Billion", var)
    return int(data)
````

````python
data["Cost"] = data["Census Cost"].apply(clean_cost)
#data['Cost'] = data.to_numeric(data['Cost'])

data
````


> ## cleaning the Average Cost




````python
def clean_avg_cost(data):
    data = data.replace("$","")
    if "cents" in data:
        data = data.replace("cents", "")
        var = float(data)/100
        data = round(var,3)
    return float(data)
````

````python
data['Avg_Cost_per_person'] = data['Average Cost Per Person'].apply(clean_avg_cost)
data
````
